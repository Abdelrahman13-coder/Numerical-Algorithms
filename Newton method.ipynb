```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sympy import 
from sympy.plotting import plot3d
```

```python
f1 = lambda x: x**2 - 15*x + 1
f1_dash = lambda x: 2*x - 15 
```

```python
def newton_single(x0,x1,tol = 0.1, iterations = 10):
    
    for i in range(iterations):
        f_d_dash = f1_dash(x1) - f1_dash(x0) / (x1-x0)
        x_new = x1 - f1_dash(x1) / f_d_dash
        if np.abs(f1_dash(x1)) < tol:
            print("Stopped at iterations {} ".format(i+1))
            break
        x0 = x1
        x1 = x_new
    return x_new
```

```python
r0 = newton_single(4.5,5)
r0
```

```python
x,y = symbols('x y')
f = x**2 + y**2
plot3d(f,size=(10,7));
```

```python
f2 = lambda x : x[0]**2 + x[1]**2
f2_gradient = lambda x : np.array([[2*x[0][0]],
                                   [2*x[1][0]]])
hessian = np.array([[2,0],
                    [0,2]])

x = np.array([[10],
              [10]])
```

```python
def newton_multi(x,hessian, tol = 0.001, iterations = 100,learning_rate = 1):
#     x_new,y_new = f2(x,y)
#     vector = np.array([[],
#                        []])
    for i in range(iterations):
        x = x - (np.linalg.inv(hessian) @ f2_gradient(x)*learning_rate)
        if np.linalg.norm(f2_gradient(x)) < tol:
            print("Stopped at iteration {}".format(i+1))
            break
    return x 
```

```python
r2 = newton_multi(x, hessian)
r2
```
